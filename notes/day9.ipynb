````python
# Let's try creating a map of the overall states...

def createMergedStates(initial_state):
    stateBuffer = [initial_state]
    mergedStates = {}
    while stateBuffer:
        state = stateBuffer.pop()
        if state in mergedStates:
            continue

        lhs, rhs = get_next_states(state)
        mergedStates[state] = (lhs, rhs)
        if lhs not in mergedStates:
            stateBuffer.append(lhs)
        if rhs not in mergedStates:
            stateBuffer.append(rhs)
    
    return mergedStates
   
````

````python
def explore_tail(states_of_interest):
    previous_states = [
        (n, v[0], v[1]) for (n, v)
        in nodeMap.items()
        if v[0] in states_of_interest or v[1] in states_of_interest
    ]
    possible_left = [(n, l, r) for (n, l, r) in previous_states if l in states_of_interest]
    possible_right = [(n, l, r) for (n, l, r) in previous_states if r in states_of_interest]
    
    possible_precursors = []
    
    if len(possible_left) == len(states_of_interest):
        left = frozenset(n for (n, _, _) in possible_left)
        possible_precursors.append(('L', left))

    if len(possible_left) > len(states_of_interest):
        print(f'uh oh: left has more states: {possible_left}')

    if len(possible_right) == len(states_of_interest):
        right = frozenset(n for (n, _, _) in possible_right)
        possible_precursors.append(('R', right))

    if len(possible_right) > len(states_of_interest):
        print(f'uh oh: right has more states: {possible_right}')
    
    return possible_precursors
````
